```python
process_values = df['processPattern'].unique()

random_samples = []

for value in process_values:
    sample = df[df['processPattern'] == value].sample(frac=0.05)
    random_samples.append(sample)

random_samples_df = pd.concat(random_samples, ignore_index=True)

random_samples_df.head(10)
```

```python
random_sample_count = random_samples_df['processPattern'].value_counts()
```

```python
df['processPattern_string'] = df['processPattern'].replace({11: "의문 유발형(부호)", 12 : '의문 유발형(은닉)',
                                                           13 : "선정표현 사용형", 14 : "속어/줄임말 사용형",
                                                           15: "사실 과대 표현형", 16 : "의도적 주어 왜곡형",
                                                           21 : "상품 판매정보 노출 광고형", 22: "부동산 판매정보 노출 광고형",
                                                           23 : "서비스 판매정보 노출 광고형", 24 : "의도적 상황 왜곡/전환형"})
```

```python
df['processPattern_string'].value_counts()
```

```python
df['useType_string'] = df['useType'].replace({0: "낚시성기사", 1 : '정상기사'})
```

```python
df['useType_string'].value_counts()
```

```python
dataset_data = [
    {
        "input" : f"기사제목: f{row_dict['newsTitle']} 기사내용: f{row_dict['newsContent']}",
        "output" : f"해당 기사는  {row_dict['useType_string']}입니다. 낚시기사 유형은 {row_dict['processPattern_string']} 입니다."  
    }
    for row_dict in df.to_dict(orient = "records")
]

dataset_data[-5:]
```
